```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1["Precio"] = pd.to_numeric(tienda1["Precio"], errors="coerce")
tienda2["Precio"] = pd.to_numeric(tienda2["Precio"], errors="coerce")
tienda3["Precio"] = pd.to_numeric(tienda3["Precio"], errors="coerce")
tienda4["Precio"] = pd.to_numeric(tienda4["Precio"], errors="coerce")

suma1 = tienda1["Precio"].sum()
suma2 = tienda2["Precio"].sum()
suma3 = tienda3["Precio"].sum()
suma4 = tienda4["Precio"].sum()

suma_total = suma1 + suma2 + suma3 + suma4

print("Suma tienda 1:", suma1)
print("Suma tienda 2:", suma2)
print("Suma tienda 3:", suma3)
print("Suma tienda 4:", suma4)
print("Suma total combinada:", suma_total)
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd


tienda1["Calificación"] = pd.to_numeric(tienda1["Calificación"], errors="coerce")
tienda2["Calificación"] = pd.to_numeric(tienda2["Calificación"], errors="coerce")
tienda3["Calificación"] = pd.to_numeric(tienda3["Calificación"], errors="coerce")
tienda4["Calificación"] = pd.to_numeric(tienda4["Calificación"], errors="coerce")

promedio1 = tienda1["Calificación"].mean()
promedio2 = tienda2["Calificación"].mean()
promedio3 = tienda3["Calificación"].mean()
promedio4 = tienda4["Calificación"].mean()

print("Promedio de calificación en Tienda 1:", round(promedio1, 2))
print("Promedio de calificación en Tienda 2:", round(promedio2, 2))
print("Promedio de calificación en Tienda 3:", round(promedio3, 2))
print("Promedio de calificación en Tienda 4:", round(promedio4, 2))
```

# 4. Productos más y menos vendidos

```python
import pandas as pd


def analizar_tienda(df, nombre):
    print(f"\n Resultados para {nombre}:")

    categoria_counts = df["Categoría del Producto"].value_counts()
    print("\nCantidad de productos vendidos por categoría:")

    producto_counts = df["Producto"].value_counts()
    mas_vendido = producto_counts.idxmax()
    menos_vendido = producto_counts.idxmin()
    print(f"\nProducto más vendido: {mas_vendido} ({producto_counts.max()} ventas)")
    print(f"Producto menos vendido: {menos_vendido} ({producto_counts.min()} ventas)")

analizar_tienda(tienda1, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tiendas = {
    "Tienda 1": pd.read_csv(url1),
    "Tienda 2": pd.read_csv(url2),
    "Tienda 3": pd.read_csv(url3),
    "Tienda 4": pd.read_csv(url4)
}

for nombre, df in tiendas.items():
    print(f"\n {nombre}")

    categorias = df["Categoría del Producto"].value_counts()
    print("\nCantidad de productos por categoría:")
    print(categorias)

    productos = df["Producto"].value_counts()

    mas_vendido = productos.idxmax()
    mas_vendido_cant = productos.max()

    menos_vendido = productos.idxmin()
    menos_vendido_cant = productos.min()

    print(f"\nProducto más vendido: {mas_vendido} ({mas_vendido_cant} unidades)")
    print(f"Producto menos vendido: {menos_vendido} ({menos_vendido_cant} unidades)")

    plt.figure(figsize=(6, 6))
    categorias.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'{nombre} - Categoría del Producto')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
import pandas as pd

def analizar_tienda(df, nombre):
    print(f"\n Resultados para {nombre}:")

    df["Costo de envío"] = pd.to_numeric(df["Costo de envío"])
    promedio_envio = df["Costo de envío"].mean()
    print(f"\nPromedio del costo de envío: ${promedio_envio:.2f}")

analizar_tienda(tienda1, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

def obtener_promedio_envio(df):
    df["Costo de envío"] = pd.to_numeric(df["Costo de envío"])
    return df["Costo de envío"].mean()

promedios_envio = {
    "Tienda 1": obtener_promedio_envio(tienda1),
    "Tienda 2": obtener_promedio_envio(tienda2),
    "Tienda 3": obtener_promedio_envio(tienda3),
    "Tienda 4": obtener_promedio_envio(tienda4),
}

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(promedios_envio.keys(), promedios_envio.values(), color='skyblue')
plt.title("Promedio de Costo de Envío por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de Envío Promedio ($)")
plt.ylim(0, max(promedios_envio.values()) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
